```python
# Importing Neccessary Libraries
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer # convert textual data into numerical values
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Data loading and preprocessing
movie_df = pd.read_csv('movies.csv')

movie_df.head()
```

```python
# Selecting The Relevant Features
sel_features = ['genres','keywords','tagline','cast','director']
```

```python
# Filling Null Values
for feature in sel_features:
    movie_df[feature] = movie_df[feature].fillna('')
```

```python
# Concetenating Selected Features
concate_features = movie_df[sel_features[0]] + " " + movie_df[sel_features[1]] + " " + movie_df[sel_features[2]] + " " + movie_df[sel_features[3]] + " " + movie_df[sel_features[4]]

print(concate_features)
```

```python
# Converting Text Data To Feature Vectors
vectorizer = TfidfVectorizer()

feature_vectors = vectorizer.fit_transform(concate_features)

print(feature_vectors)
```

```python
# Similarity Confidence Value
cos_similarity = cosine_similarity(feature_vectors)

print(cos_similarity)
```

```python
# User Input
user_input = input('Enter The Movie You Want To Search : ')
```

```python
movie_names = movie_df['title'].tolist()

print(movie_names)
```

```python
close_match = difflib.get_close_matches(user_input, movie_names)
print(close_match)
```

```python
movie_index = movie_df[movie_df.title == close_match[0]]['index'].values[0]
print(movie_index)
```

```python
# Retrieving A List Of Similar Movies
sim_score = list(enumerate(cos_similarity[movie_index]))
print(sim_score)
```

```python
# Sorting Movies Based On Score
sorted_sim_score = sorted(sim_score, key=lambda x:x[1], reverse=True)
print(sorted_sim_score)
```

```python
# Extracting Movies Based On Index
print("Suggested Movies : ")

i=1

for movies in sorted_sim_score:
    index = movies[0]
    title_from_index = movie_df[movie_df.index == index]['title'].values[0]
    if i < 30:
        print(i, '- ', title_from_index)
    i += 1
```
